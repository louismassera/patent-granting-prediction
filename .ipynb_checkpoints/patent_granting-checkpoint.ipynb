```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
import datetime
from sklearn.preprocessing import Imputer, OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
```

# Linear Model

```python
df_train = pd.read_csv('data/train.csv', sep=';')
df_test = pd.read_csv('data/test.csv', sep=';')
y_train = (df_train['VARIABLE_CIBLE'] == 'GRANTED').astype(int).ravel()
df_train.drop('VARIABLE_CIBLE', axis=1, inplace=True)
df_concat = pd.concat((df_train, df_test), axis=0, ignore_index=True)
```

```python
print("Patent granted proportion =",y_train.sum()/len(y_train))
```

```python
def imputer(df):
    df = df.copy()
    features_num = list(df.columns[df.dtypes != 'object'])
    features_cat = list(df.columns[df.dtypes == 'object'])

    for i in features_cat:
        df.loc[df[i].isnull(), i] = "value_missing"
        
    df[features_cat] = df[features_cat].apply(LabelEncoder().fit_transform)
    df[features_num] = Imputer().fit_transform(df[features_num])
    
    return df
```

```python
def plot_important_features(coef, feature_names, top_n=20, ax=None, rotation=60):
    if ax is None:
        ax = plt.gca()
    inds = np.argsort(coef)
    low = inds[:top_n]
    high = inds[-top_n:]
    important = np.hstack([low, high])
    myrange = range(len(important))
    colors = ['red'] * top_n + ['blue'] * top_n
    ax.bar(myrange, coef[important], color=colors)
    ax.set_xticks(myrange)
    ax.set_xticklabels(feature_names[important], rotation=rotation, ha="right", fontsize=14)
    ax.set_xlim(-.7, 2 * top_n)
    ax.set_frame_on(False)
```

```python
X_train = imputer(df_train)
```

```python
%%time
param_grid = {'logistic__C': np.logspace(-4,4,10)}
linear_pipe = Pipeline([('scaler', StandardScaler()),
                        ('logistic', LogisticRegression())])
linear_grid = GridSearchCV(linear_pipe, 
                           param_grid=param_grid,
                           scoring = 'roc_auc',
                           cv=5,
                           verbose=True)
linear_grid.fit(X_train, y_train)
```

```python
plt.figure(figsize=(16,5))
plot_important_features(linear_grid.best_estimator_.steps[1][1].coef_[0], X_train.columns, top_n=20)
plt.show()
```

# Feature Engineering

```python
dates = ['PRIORITY_MONTH', 'FILING_MONTH', 'PUBLICATION_MONTH', 'BEGIN_MONTH']
features_num = list(df_concat.columns[df_concat.dtypes != 'object']) + dates
features_cat = list(set(list(df_concat.columns[df_concat.dtypes == 'object'])) - set(dates))
label_encoded = list(set(features_cat) - set(dates))
one_hot_encoded = list(set(label_encoded) - set(['MAIN_IPC', 'FIRST_CLASSE']))

final_features = dates + features_num + ['MAIN_IPC', 'FIRST_CLASSE']
```

```python
def convert_date(date):
    if date is np.nan:
        return None
    return time.mktime(datetime.datetime.strptime(date, "%m/%Y").timetuple())
```

```python
for i in dates:
    df_concat[i] = df_concat[i].map(convert_date)
df_concat['TIME_DIFF'] = (df_concat['FILING_MONTH'] - df_concat['PRIORITY_MONTH'])/(3600*24)

for i in features_cat:
    df_concat.loc[df_concat[i].isnull(), i] = "value_missing"
df_concat[label_encoded] = df_concat[label_encoded].apply(LabelEncoder().fit_transform)
X_encoded = OneHotEncoder(sparse=False).fit_transform(df_concat[one_hot_encoded])
df_concat[features_num] = Imputer().fit_transform(df_concat[features_num])

df_train = df_concat[:len(df_train)][final_features]
df_test = df_concat[len(df_train):][final_features]
X_encoded_train, X_encoded_test = X_encoded[:len(df_train)], X_encoded[len(df_train):]
X_train = np.concatenate((df_train.as_matrix(), X_encoded_train),axis=1)
X_test = np.concatenate((df_test.as_matrix(), X_encoded_test), axis=1)
```

# Grid-search

```python
param_grid={'n_estimators':[50,100,150,200],
            'max_depth': [2,4,8]}
grad_grid = GridSearchCV((GradientBoostingClassifier(max_features=0.3),
                          param_grid=param_grid,
                          scoring='roc_auc',
                          n_jobs=-1,
                          cv=5,
                          verbose=True)
```

# Final model and prediction

```python
%%time
clf = GradientBoostingClassifier(n_estimators=150,max_depth = 8, max_features=0.3)
clf.fit(X_train, y_train)
y_pred = clf.predict_proba(X_test)
np.savetxt('y_pred.txt', y_pred[:,1], fmt='%s')
```
